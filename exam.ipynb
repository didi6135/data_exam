```python
br_data = pd.read_csv("./exam/BR.csv")
fedfunds_data = pd.read_csv("./exam/FEDFUNDS.csv")
usd_ils_data = pd.read_csv("./exam/USD_ILS.csv")

usd_ils_data.head(5) 

```

```python
df_israel = pd.DataFrame(br_data)
df_israel['Year'] = pd.to_datetime(df_israel['TIME_PERIOD']).dt.year
df_israel['Month'] = pd.to_datetime(df_israel['TIME_PERIOD']).dt.month

# Rename 'OBS_VALUE' to 'Current_Rate' (adjust 'OBS_VALUE' if the column name differs)
df_israel = df_israel.rename(columns={'OBS_VALUE': 'Current_Rate'})

# Calculate the change from the previous month
df_israel['Change_From_Last_Month'] = df_israel['Current_Rate'].diff()

# Optional: Calculate Change_Percentage
df_israel['Change_Percentage'] = df_israel['Change_From_Last_Month'] / df_israel['Current_Rate'].shift(1) * 100

# Select only relevant columns
israel_interest_rates = df_israel[['Year', 'Month', 'Current_Rate', 'Change_From_Last_Month', 'Change_Percentage']]

# israel_interest_rates
```

```python
df_fedfunds = pd.DataFrame(fedfunds_data)

# Convert 'DATE' to datetime format and extract Year and Month
df_fedfunds['Year'] = pd.to_datetime(df_fedfunds['DATE']).dt.year
df_fedfunds['Month'] = pd.to_datetime(df_fedfunds['DATE']).dt.month

# Rename 'FEDFUNDS' to 'Current_Rate'
df_fedfunds = df_fedfunds.rename(columns={'FEDFUNDS': 'Current_Rate'})

# Calculate Change_From_Last_Month
df_fedfunds['Change_From_Last_Month'] = df_fedfunds['Current_Rate'].diff()

# Optional: Calculate Change_Percentage
df_fedfunds['Change_Percentage'] = df_fedfunds['Change_From_Last_Month'] / df_fedfunds['Current_Rate'].shift(1) * 100

# Select relevant columns
fedfunds_processed = df_fedfunds[['Year', 'Month', 'Current_Rate', 'Change_From_Last_Month', 'Change_Percentage']]

# Display the processed DataFrame
# fedfunds_processed
```

```python
df_exchange = pd.DataFrame(usd_ils_data)

# Convert 'תאריך' to datetime format and extract Year and Month
df_exchange['תאריך'] = pd.to_datetime(df_exchange['תאריך'], format='%d.%m.%Y')
df_exchange['Year'] = df_exchange['תאריך'].dt.year
df_exchange['Month'] = df_exchange['תאריך'].dt.month

# Rename 'שער' to 'Current_Rate'
df_exchange = df_exchange.rename(columns={'שער': 'Current_Rate'})

# Sort data by date to calculate month-to-month changes
df_exchange = df_exchange.sort_values(by='תאריך').reset_index(drop=True)

# Calculate Change_From_Last_Month based on sorted data
df_exchange['Change_From_Last_Month'] = df_exchange['Current_Rate'].diff()

# Optional: Calculate Change_Percentage from the rate difference
df_exchange['Change_Percentage'] = df_exchange['Change_From_Last_Month'] / df_exchange['Current_Rate'].shift(1) * 100

# Select only relevant columns
exchange_processed = df_exchange[['Year', 'Month', 'Current_Rate', 'Change_From_Last_Month', 'Change_Percentage']]

# Display the processed DataFrame
exchange_processed
```

```python
# Define event periods based on the recommended events
events = {
    "US_Rate_Hike_Dec_2015": ("2015-12-01", "2016-02-29"),
    "Israel_Rate_Cut_Jan_2015": ("2015-01-01", "2015-03-31"),
    "Israel_Negative_Rate_Discussion_2015": ("2015-02-01", "2015-04-30")
}

# Store results
event_analyses = {}

# Analyze each event
for event_name, (start_date, end_date) in events.items():
    # Filter data for the event period based on Year and Month columns
    event_data = all_data_merged[
        (all_data_merged['Year'].astype(str) + '-' + all_data_merged['Month'].astype(str).str.zfill(2)).between(start_date, end_date)
    ]
    
    # Calculate average Rate_Difference and Current_Rate in this period
    avg_rate_diff = event_data['Rate_Difference'].mean()
    avg_exchange_rate = event_data['Current_Rate'].mean()
    
    # Calculate correlation between Rate_Difference and Current_Rate for this period
    event_correlation = event_data['Rate_Difference'].corr(event_data['Current_Rate'])
    
    # Store results
    event_analyses[event_name] = {
        "average_rate_difference": avg_rate_diff,
        "average_exchange_rate": avg_exchange_rate,
        "event_correlation": event_correlation,
        "event_data_sample": event_data.head()
    }

# Display results with explanations (in Hebrew)
for event, analysis in event_analyses.items():
    print(f"אירוע: {event}")
    print(f"השפעה על ההפרש בין הריביות: {analysis['average_rate_difference']}")
    print(f"שיעור חליפין ממוצע: {analysis['average_exchange_rate']}")
    print(f"קורלציה בין ההפרש לריבית במהלך האירוע: {analysis['event_correlation']}")
    # print("דוגמה מהנתונים באירוע זה:")
    # print(analysis['event_data_sample'])
    print("\n")
```
